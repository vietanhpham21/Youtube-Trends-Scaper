Skript dass in den Datensätzen im Verzeichnis "scraped_data" schaut und alle Datein durchsucht. Es werden alle uniquen Channels extra gespeichert in "unique_channels"

Identifiziert alle einzigartigen Channels mittels Channel Url. wenn sie nicht in der "unique_channels" datei ist wird der kanal hinzugefügt

```python
import os
import json

def save_unique_channels(data, unique_channels_file):
    unique_channels = {}

    try:
        if os.path.exists(unique_channels_file):
            with open(unique_channels_file, 'r', encoding='utf-8') as file:
                unique_channels = json.load(file)
    except Exception as e:
        raise Exception(f"Error loading {unique_channels_file}: {str(e)}")
    
    # Füge neue Kanäle hinzu
    for video in data.get('videos', []):
        channel_url = video.get('channel_url')
        if channel_url and channel_url not in unique_channels:

            unique_channels[channel_url] = {
                'channel_name': video.get('channel'),
            }

    # aktualisieren der Datei
    try:
        with open(unique_channels_file, 'w', encoding='utf-8') as file:
            json.dump(unique_channels, file, ensure_ascii=False, indent=4)
            print(f'Unique channels saved to "{unique_channels_file}"')
    except Exception as e:
        raise Exception(f"Error saving {unique_channels_file}: {str(e)}")
```

Durchläuft den "folder_path"  und ruft für jede datei die Funktion "save_unique_channels" auf

```python
folder_path = 'scraped_data/'

for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        print(f'Found file: {file_path}')

        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                unique_channels_file = 'unique_channels/unique_channels.json'
                save_unique_channels(data, unique_channels_file)
        except Exception as e:
            print(f"Error processing {file_path}: {str(e)}")
```
